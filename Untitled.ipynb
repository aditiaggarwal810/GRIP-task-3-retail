```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

## State Analysis

```python
state=df.groupby('State')['Quantity'].sum().reset_index()
state=state.sort_values(by=['Quantity'],ascending=False)
state
```

```python
plt.figure(figsize=(20,20))
sns.set(font_scale=1.6)
sns.barplot(x='State',y='Quantity',data=state)
plt.xticks(rotation=90)
plt.show()
```

```python
df_state=df.groupby('State')[['Sales','Profit','Discount']].sum().reset_index()
df_state=df_state.sort_values(by=['Profit'],ascending=False)
df_state
```

```python
plt.figure(figsize=(20,20))
sns.set(font_scale=1.6)
sns.barplot(x='State',y='Sales',data=df_state)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.set(font_scale=1.6)
sns.barplot(x='State',y='Profit',data=df_state)
plt.xticks(rotation=90)
plt.show()
```

## From this it s very clear , California is doing amazing among all the states. California , Newyork, Washington and Michigan are doing good while Texas, Ohio, Pennysylvania and Illinios are doing bad in terms of profit while doing good sales.

## Region Analysis

```python
region=df.groupby('Region')['Quantity'].count().reset_index()
region
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Quantity',data=region,color='Violet',edgecolor='black')
plt.show()
```

```python
df_region=df.groupby('Region',as_index=False)[['Sales','Profit']].sum()
df_region
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.bar(df_region['Region'],df_region['Sales'],color='red')
plt.bar(df_region['Region'],df_region['Profit'],color='Green')

plt.subplot(1,2,2)
plt.pie(df_region['Profit'],autopct='%1.1f%%',labels=df_region['Region'],explode=[0,0,0,0.3],shadow=True)
plt.show()
```

## West region is getting maximum profit while east is also doing good but central needs to do good . South is getting good profit as compare to its sales.

## Segment analysis

```python
segment=df['Segment'].value_counts().reset_index()
segment
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='index',y='Segment',data=segment,color='blue',edgecolor='black')
plt.xlabel("Segment")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='Region',y='Profit',hue='Segment',data=df,palette='bright')
plt.show()
```

```python
df_segment=df.groupby('Segment',as_index=False)[['Profit','Sales']].sum()
df_segment
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.bar(df_segment['Segment'],df_segment['Sales'],color='red')
plt.bar(df_segment['Segment'],df_segment['Profit'],color='Green')

plt.subplot(1,2,2)
plt.pie(df_segment['Profit'],autopct='%1.1f%%',labels=df_segment['Segment'],explode=[0.3,0,0],shadow=True)
plt.show()
```

## Consumer segment is doing better than corporate and home office.
